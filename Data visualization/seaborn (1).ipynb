```python
import seaborn as sns

sns.set_theme()

tips=sns.load_dataset("tips")


```

```python
tips.head()
```

```python
tips.info()
```

```python

tips.day.value_counts(1)
tips.sex.value_counts(1)
```

```python
ax=sns.relplot(
    data=tips,
    x="total_bill",y="tip",col="time",hue='smoker',size='size',style='smoker'
)
ax.set(xlabel="xlabel")
```

```python
fmri=sns.load("fmri")
sns.relplot(
    data=fmri,kind='line',
    x="timepoint",y="signal", col="region",hue="event" ,style="event",
)
```

```python
sns.lmplot(data=tips,x="total_bill",y="tip",col="time",hue="smoker")
```

```python
sns.displot(data=tips,x="total_bill",col="time",kde=True)
```
